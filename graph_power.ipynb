```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("resultsboth/gpg22-10s.csv", sep=',')
grouped = df.groupby('load')
```

```python
for name, g in grouped:
    plt.scatter(g['util'], g['energy'])
    plt.xlabel('CPU Utilization (%)')
    plt.ylabel('Power Usage (Watts)')
    plt.title(f'CPU Utilization vs Power Usage ({name}% load per thread)')
    plt.axvline(800)
    plt.axvline(1600)
    plt.axvline(3200)
    plt.show()
```
